```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def enroll_course(self):
        print(f"Student {self.name} is enrolling in a course.")

    # Overriding display_info as per requirement #5
    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")

class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self):
        print(f"Lecturer {self.name} is teaching a course.")

# Demonstration
# Creating objects
student1 = Student("Fred", 20, "S12345")
lecturer1 = Lecturer("Mr. Matthew Cobbinah", 45, "E98765")


student1.display_info()
student1.enroll_course()

print("-" * 20)

lecturer1.display_info()
lecturer1.teach_course()
```
